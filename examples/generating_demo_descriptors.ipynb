```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("biosieve_example_dataset_1000.csv")
```

```python
rng = np.random.default_rng(13)
for i in range(32):  
    df[f"desc_{i:03d}"] = rng.normal(size=len(df)).astype("float32")
df.to_csv("biosieve_example_dataset_1000_desc.csv", index=False)
```
